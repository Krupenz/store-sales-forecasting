```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('train.csv')
dataset['date'] = pd.to_datetime(dataset['date'], infer_datetime_format=True)
dataset = dataset.set_index(['id'])
#dataset = dataset.sort_values(["date", "store_nbr"], ascending = (True, True))
print(dataset)
```

```python
#stworzenie dicta który zawiera dane podzielone na sklepy i kategorie; nazwy keysów: f"{store}{family}"
#odpalenie tej komórki trwa kilka minut takze cierpliwosci
stores = dataset['store_nbr'].drop_duplicates()
families = dataset['family'].drop_duplicates()
print(len(stores))
print(len(families))
dataset_dict = {}
for store in stores:
    for family in families:
        #print(f"{store} {family}")
        temp_df = dataset.loc[(dataset['store_nbr'] == store) & (dataset['family'] == family)]
        dataset_dict[f"{store}{family}"]=temp_df
```

```python
print(dataset_dict)
```

```python
#zgrupowanie danych po dniu roku
temp_df = dataset_dict["10BABY CARE"]
print(temp_df)
temp_df=temp_df.loc[pd.to_datetime(temp_df['date']).dt.dayofyear == 1]
print(temp_df)
```
